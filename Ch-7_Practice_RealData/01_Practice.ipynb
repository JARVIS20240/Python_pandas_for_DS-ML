Q2 Find:

Total missing values per column, 
Percentage missing, 
Column with highest null ratio

```python
# Total Missing Walue per Column:
print(df.isnull().sum())
```

```python
# Percentage Missing Value:
# null_p = (df.isnull().mean()*100).round(2) #round for how many numbers after (.) 
null_per = (df.isnull().sum()/ len(df) * 100).round(2)
print(null_per)
```

```python
# Column with highest null ratio
# find Null Ration:
null_ratio = (df.isnull().mean()).round(2)
print(null_ratio)
```

```python
# finding Max Null Ratio:
Max_null_ratio = null_ratio.idxmax()
print(Max_null_ratio, max(null_ratio))
```

Drop columns with >40% missing, Fill numeric nulls using median, Fill categorical nulls using mode

```python
print(null_per[null_per > 40])
columns_grater_40_per = null_per[null_per>40].index #to find Index/Lable with specific values
print(columns_grater_40_per)
```

```python
# Drop Columns with >40%:
df = df.drop(columns=columns_grater_40_per)
```

```python
# Fill numeric nulls using median

print(df.isnull().sum())
print((df.isnull().sum()/len(df) * 100).round(2))
```

```python
numeric_cols = df.select_dtypes(include="number")
# df[numeric_cols.columns] = numeric_cols.fillna(numeric_cols.mean())
df.fillna(df.mean(numeric_only=True), inplace=True)
df.isnull().sum()
```
