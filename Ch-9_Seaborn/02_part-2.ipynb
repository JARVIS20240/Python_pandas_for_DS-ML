#### Categorical Scatter Plot

- Stripplot

```python
# in Normal we Plot a Scatter plot on 2-Numerical field
# Normal scatter :
sns.scatterplot(data=tips, x="total_bill", y="tip")
```

- Stripplot :

```python
# stripplot : categorical and Numerical Scatter plot
sns.stripplot(data=tips, x="day", y="total_bill", jitter=False, hue="sex") #axis level function
# jitter = removie noice from plot, data in single line
```

```python
# Figure Level : catplot
sns.catplot(data=tips, x="day", y="total_bill", kind="strip", hue="sex")
```

```python
# jitter :
sns.catplot(data=tips, x="day", y="total_bill", kind="strip", hue="sex", jitter = 0.2, palette='dark:red')
```

- Swarmplot

```python
sns.catplot(data=tips, x="day", y="total_bill", kind="swarm", hue="sex")
```
