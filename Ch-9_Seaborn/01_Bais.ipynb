#### 1. Relational Plot

- to see the statistical relation between 2 or more variables.
- Bivariate Analysis

Plots under this section

- scatterplot
- lineplot

#### Axis Level -> scatterplot, lineplot
#### Figure Leven -> relplot 

```python
import numpy as np
import pandas as pd
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
```

- Scatter Plot :

```python
tips = sns.load_dataset("tips")
tips.head(2)
```

```python
# Scatterplot: -> Axis Level Function

sns.scatterplot(data=tips, x='total_bill', y="tip", hue="sex", style="time", size="size")
# hue = "can Seprate Data with Male and Female here"
# style = "Seprate data with time here"
# size = "Seperate data with size here customer total size"
```

```python
# relplot: -> Fegure Level Function
sns.relplot(data=tips, x='total_bill', y="tip", kind="scatter", hue="sex", style="time", size="size")
```

- Line Plot

```python
gap = px.data.gapminder()
gap.head(2)
india_df = gap[gap['country'] == "India"]
india_df.head(2)
```

```python
# lineplot: -> Axis Level Function
sns.lineplot(data=india_df, x="year",y="pop")
```

```python
# relplot: -> Figure Level Function
sns.relplot(data=india_df, x="year", y="pop", kind="line")
```

```python
temp_df = gap[gap['country'].isin(['India', 'Brazil', 'Germany'])]
temp_df.head()
```

```python
# hue -> style
sns.relplot(kind="line", data=temp_df, x="year", y="lifeExp", hue='country', style="continent", size="continent")
```

##### Facet Plot : 
- we makes multiple plots with one categorical col
- work only Figure level Function (sns.relplot()) but not work with Axis Level plot

```python
# side by side plot:

sns.relplot(data=tips, x="total_bill", y="tip", kind="scatter",row='day', col="sex")
# col = here col make each Catagorical <Value seperated Scattre Plot (column wise side by side)
# row = here col make each Catagorical <Value seperated Scattre Plot (row wise side by side)
```

```python
gap.sample(3)
```

```python
# Col Wrap
sns.relplot(data=gap, x="lifeExp", y="gdpPercap", kind="scatter", col="year", col_wrap=3)
# col_warp: only 3 graph in one row
```

##### 2. Distribution Plots

- used for univariate analysis (single col)
- used to find out the distribution
- Range of the observation
- Central Tendency
- is the data bimodal?
- Are there outliers?

Plots under distribution plot

- histplot
- kdeplot
- rugplot

#### Axis Level -> histplot, kdeplot, rugplot 
#### Fegure Level -> displot

- Histplot:

```python
tips.head()
```

```python
# histplot:
sns.histplot(data=tips, x="total_bill")
```

```python
# displot:
sns.displot(data=tips, x="total_bill", kind="hist")
```

```python
# bins parameters:
sns.displot(data=tips, x="total_bill", kind="hist", bins = 20)
# bins = total bins on graph
```

```python
# histogram on Categorical data:
sns.displot(data=tips, x="day", kind="hist")

# kind of Count plot of categorical data
```

```python
# hue
sns.displot(data=tips, x="tip", kind="hist", hue="sex")
```

```python
# elements = "steps"
sns.displot(data=tips, x="tip", kind="hist", hue="sex", element="step")
# element = "step" only provide outlines of histtogram
```

```python
titanic = sns.load_dataset("titanic")
titanic.head(2)
```

```python
# faceting using col & row -> this not work on Axis level Function 
sns.displot(data=titanic, x="age", kind="hist", hue="survived", col="sex")
```

#### kdeplot :
- KDE = Kernel Density Estimation

    - use Gaussion Kernal, and make Data Distibution smuther then Histplot

```python
# kdeplot :
sns.kdeplot(data=tips, x="total_bill")
```

```python
# displot :
sns.displot(data=tips, x="total_bill", kind="kde")
```

```python
# hue :
sns.displot(data=tips, x="total_bill", kind="kde", hue="sex", fill=True, height=5, aspect=2)
```

#### Rugplot

- Plot marginal distributions by drawing ticks along the x and y axes.
- This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.

```python
sns.kdeplot(data=tips, x="total_bill")
sns.rugplot(data=tips, x="total_bill")
```

- Bivariate histogram :


```python
# Bivariate Histplot
sns.displot(data=tips, x="total_bill", y="tip", kind="hist")
# darker shade in Plot hase both Data match or More data in that part 
```

```python
# Bivariate Kdeplot
sns.displot(data=tips, x="total_bill", y="tip", kind="kde")
# Closest lines have multiple data at that point
```

#### Metrix Plot:
- Heatmap
- clustermap

```python
# Heatmap : heatmap is Axis level Function , No figure level function 
df = gap.pivot(index="country", columns="year", values="lifeExp")
df.head()
```

```python
plt.figure(figsize=(16,12))
sns.heatmap(data=df)
```

```python
temp_df = gap[gap['continent'] == "Europe"].pivot(index="country", columns="year", values="lifeExp")
temp_df.head()
```

```python
plt.figure(figsize=(16,12))
sns.heatmap(temp_df, annot=True, linewidths=0.9, cmap="crest")
# annot = True -> for Display numbers/values
# linewidth = 0.5
# cmap for colors -> viridis, plasma, inferno, magma, crest etc....
```

```python
# Clustermap
iris = px.data.iris()
iris.head()
```

```python
# clustermap
sns.clustermap(iris.iloc[:,[ 0,1,2,3]])
```
